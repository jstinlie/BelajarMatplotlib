```python
# import matplotlib, matplotlib.pyplot, dan numpy
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
# simple line plot
x = np.arange(0.0, 4.0, 0.01)
x
```

```python
# menampung nilai hasil operasi sin dari data yang di tampung di varable x
s = np.sin(2 * np.pi * x)
s
```

```python
# plotting dari dua data set yang dimiliki
fig, ax = plt.subplots() # OO Style

ax.plot(x, s)
ax.set(xlabel='nilai x',
       ylabel='nilai sin(x)',
       title='Visualisasi nilai sin')
ax.grid()

plt.show()
```

```python
# plotting dari dua set yang dimiliki
plt.plot(x, s) # pyplot style

plt.xlabel('nilai x')
plt.ylabel('nilai sin(x)')
plt.title('Visualisasi nilai sin')
plt.grid()
```

```python
# multiple subplots
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
x1
```

```python
# multiple subplots dengan OO Style
fig, (ax1, ax2) = plt.subplots(2,1)
fig.suptitle('Multiple SUbplots')

ax1.plot(x1, y1, 'ro-')
ax1.set_ylabel('Nilai $cos(x_1)$')

ax2.plot(x2, y2, 'g.-')
ax2.set_ylabel('Nilai $cos(x_2)$')

ax2.set_xlabel('Nilai x')

plt.show()
```
