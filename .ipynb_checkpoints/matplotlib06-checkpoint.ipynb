```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
# simple pie plot
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 140, 245, 210, 225]
```

```python
plt.pie(data,
        labels=kategori,
        autopct='%1.1f%%',
        startangle = 90)

plt.title('Contoh Pie Plot')

plt.show()
```
