```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
# simple pie plot
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 140, 245, 210, 225]
```

```python
plt.pie(data,
        labels=kategori,
        autopct='%1.1f%%',
        startangle = 90)

plt.title('Contoh Pie Plot')

plt.show()
```

```python
# pengaturan legend dan color
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna =['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue'] 
```

```python
plt.pie(data,
        labels=kategori,
        colors=warna,
        autopct='%1.1f%%',
        startangle = 90)

plt.legend()
plt.title('Contoh Pie Plot')

plt.show()
```

```python
# exploded pie plot
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna =['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue'] 
explode_var = [0.,0.,0.2,0.,0.08]
```

```python
plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle = 90)

plt.title('Contoh Pie Plot')

plt.show()
```
