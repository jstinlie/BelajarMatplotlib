```python
# import matplotlib, matplotlib.pyploy,dan numpy
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
# membuat plotting sederhana
plt.plot([2, 4, 8, 10])
plt.ylabel('sumbu y')
plt.xlabel(' sumbu x ')
plt.show()
```

```python
plt.plot([3, 6, 7, 10], [7, 2, 6, 22]) # [3, 6, 7, 10] merupakan sumbu x, sedangkan [7, 2, 6, 22] merupakan sumbu y
plt.show()
```

```python
# pengaturan format pada plot 'bo'
plt.plot([5, 4, 3, 2], [3, 5, 10, 22], 'bo')
plt.axis([0, 7, 0, 26])
plt.show()
```

```python
# pengaturan format pada plot 'go'
plt.plot([5, 3, 2, 1], [3, 5, 15, 26], 'go')
plt.axis([0, 7, 0, 27])
plt.show()
```

```python
# pengaturan format pada plot 'ro'
plt.plot([5, 3, 2, 1], [3, 5, 15, 26], 'ro')
plt.axis([0, 7, 0, 27])
plt.show()
```

```python
# pengaturan format pada plot 'b-'
plt.plot([5, 3, 2, 1], [3, 5, 15, 26], 'b-')
plt.axis([0, 7, 0, 27])
plt.show()
```

```python
# pengaturan format pada plot 'b*'
plt.plot([5, 3, 2, 1], [3, 5, 15, 26], 'b*')
plt.axis([0, 7, 0, 27])
plt.show()
```

```python
# multi plot dalam satu pemanggilan fungsi plot()
t = np.arange(0., 10., 0.10)
t
```

```python
# multi plot dalam satu pemanggilan fungsi plot()
plt.plot(t, t, 'bo')
plt.plot(t, t**2, 'g--')
plt.plot(t, t**3, 'r*')
plt.show()
```

```python
# plotting dengan keywords 
data = {'a' : np.arange(10),
        'c' : np.random.randint(0, 5, 10),
        'd' : np.random.randn(10)}
data['b'] = data['a'] + 10 * np.random.randn(10)
data['d'] = np.abs(data['d']) * 10

data
```

```python
plt.scatter('c', 'd', data=data)
plt.xlabel('entry c')
plt.ylabel('entry d')
plt.show()
```

```python
# plotting untuk tipe data categorical
names = ['Matt', 'George', 'Luigi', 'Enix', 'Dom']
values = [50, 500, 5, 25, 125]
```

```python
plt.plot(names, values)
plt.show()
```

```python
# scatter plot
plt.scatter(names, values)
plt.show()
```

```python
# bar plot
plt.bar(names, values)
plt.show()
```
