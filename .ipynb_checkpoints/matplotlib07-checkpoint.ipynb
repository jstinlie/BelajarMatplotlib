```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
# sample data
data_uas = [
    ['Andre', 70],
    ['William', 83],
    ['Pierson', 62],
    ['Axel', 74],
    ['Matthew', 71]
]
```

```python
data_uas
```

```python
# table plot dengan OO Style
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas,
                  loc='center')

table.set_fontsize(14)
table.scale(1, 4) #ukuran kolom, baris

ax.axis(False)

plt.show()
```

```python
# table plot dengan pyplot Style
table = plt.table(cellText=data_uas,
                  loc='center')

table.set_fontsize(14)
table.scale(1, 4) # ukuran kolom, baris

ax = plt.gca()
ax.axis(False)

plt.show()
```
